```python
# Python
def ohms_law():
    print("Ohm's Law Calculator")
    while True:
        choice = input("Select calculation (V, I, R) or 'exit' to quit: ")
        if choice == 'exit':
            break
        elif choice == 'V':
            R = float(input("Enter resistance (ohms): "))
            I = float(input("Enter current (amps): "))
            V = R * I
            print(f"Voltage (V) = {V} volts")
        # Add similar calculations for I and R
        else:
            print("Invalid choice. Try again.")

ohms_law()
```

```kotlin
// Kotlin
fun ohmsLaw() {
    println("Ohm's Law Calculator")
    while (true) {
        print("Select calculation (V, I, R) or 'exit' to quit: ")
        val choice = readLine()
        if (choice == "exit") {
            break
        } else {
            when (choice) {
                "V" -> {
                    print("Enter resistance (ohms): ")
                    val R = readLine()!!.toDouble()
                    print("Enter current (amps): ")
                    val I = readLine()!!.toDouble()
                    val V = R * I
                    println("Voltage (V) = $V volts")
                }
                // Add similar calculations for I and R
                else -> println("Invalid choice. Try again.")
            }
        }
    }
}

fun main() {
    ohmsLaw()
}
```

**Week 2 Homework Assignment - Circuit Analysis (Python/Kotlin):**

```python
# Python
def total_resistance(resistors):
    return sum(resistors)

def circuit_analysis():
    print("Circuit Analysis")
    resistors = []
    while True:
        R = float(input("Enter resistance (ohms) or 0 to finish: "))
        if R == 0:
            break
        resistors.append(R)
    
    total_R = total_resistance(resistors)
    print(f"Total resistance = {total_R} ohms")

circuit_analysis()
```

```kotlin
// Kotlin
fun totalResistance(resistors: List<Double>): Double {
    return resistors.sum()
}

fun circuitAnalysis() {
    println("Circuit Analysis")
    val resistors = mutableListOf<Double>()
    while (true) {
        print("Enter resistance (ohms) or 0 to finish: ")
        val R = readLine()!!.toDouble()
        if (R == 0.0) {
            break
        }
        resistors.add(R)
    }

    val totalR = totalResistance(resistors)
    println("Total resistance = $totalR ohms")
}

fun main() {
    circuitAnalysis()
}
```
